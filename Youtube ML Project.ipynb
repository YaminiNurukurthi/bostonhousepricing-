```python
## Independent(others) and dependent(price) features

x=boston_df.iloc[:,:-1]
y=boston_df.iloc[:,-1]
```

```python
x.head()
```

```python
# train test split


from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
x_test
```

```python
y_test
```

```python
## Standardize the datset

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
# print the coefficents and intercepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

```python
# Prediction with Test Data
```

```python
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

ASSUMPTIONS

```python
# plt a scatter plot for the prediction
plt.scatter(reg_pred,y_test)
```

```python
# residuals(error)
residuals=y_test-reg_pred
```

```python
residuals
```

```python
# plotting the Residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
#scatter plot with respect to prediction and residual
```

```python
plt.scatter(reg_pred,residuals)
```

 Performance Mertics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error


print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
## performance metrics R quare and adjusted R square

from sklearn.metrics import r2_score

score=r2_score(y_test,reg_pred)
print(score)
```

```python
# Adjusted R2 Square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
